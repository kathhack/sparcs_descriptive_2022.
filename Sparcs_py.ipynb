```python
# Step 1: Calculate Average Total Cost by Age Group
average_cost_by_age_group = filtered_df_cleaned.groupby('Age_Group')['Total_Costs'].mean().reset_index()

# Step 2: Calculate Average Total Cost by Type of Admission
average_cost_by_type_of_admission = filtered_df_cleaned.groupby('Type_of_Admission')['Total_Costs'].mean().reset_index()

# Step 3: Visualize Results

# Bar plot for Average Total Cost by Age Group
plt.figure(figsize=(12, 6))
sns.barplot(x='Age_Group', y='Total_Costs', data=average_cost_by_age_group, palette="viridis")
plt.title('Average Total Cost by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Total Cost ($)')
plt.xticks(rotation=45)
plt.show()

# Bar plot for Average Total Cost by Type of Admission
plt.figure(figsize=(12, 6))
sns.barplot(x='Type_of_Admission', y='Total_Costs', data=average_cost_by_type_of_admission, palette="magma")
plt.title('Average Total Cost by Type of Admission')
plt.xlabel('Type of Admission')
plt.ylabel('Average Total Cost ($)')
plt.xticks(rotation=45)
```

```python
# noticable trends - look into age group and their average total charges and admission counts
# Step 1: Group by Age Group and calculate average total charges and admission counts
trends_by_age_group = filtered_df_cleaned.groupby('Age_Group').agg(
    Average_Charges=('Total_Charges', 'mean'),
    Admission_Count=('Type_of_Admission', 'count')
).reset_index()

# Step 2: Visualize Trends by Age Group
plt.figure(figsize=(12, 6))

# Bar plot for Average Total Charges by Age Group
plt.subplot(1, 2, 1)
sns.barplot(x='Age_Group', y='Average_Charges', data=trends_by_age_group, palette='viridis')
plt.title('Average Total Charges by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Total Charges ($)')
plt.xticks(rotation=45)

# Bar plot for Admission Count by Age Group
plt.subplot(1, 2, 2)
sns.barplot(x='Age_Group', y='Admission_Count', data=trends_by_age_group, palette='magma')
plt.title('Admission Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Admission Count')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Extra Credit

# Assuming 'filtered_df_cleaned' is your DataFrame after handling missing data

# Step 1: Calculate Average Total Charges and Length of Stay by Ethnicity and Race
ethnicity_race_analysis = filtered_df_cleaned.groupby(['Ethnicity', 'Race']).agg(
    Average_Total_Charges=('Total_Charges', 'mean'),
    Average_Length_of_Stay=('Length_of_Stay', 'mean')
).reset_index()

# Step 2: Create a Heatmap for Average Total Charges by Ethnicity and Race
plt.figure(figsize=(12, 8))
charges_pivot = ethnicity_race_analysis.pivot("Ethnicity", "Race", "Average_Total_Charges")
sns.heatmap(charges_pivot, annot=True, fmt=".2f", cmap="YlGnBu", cbar_kws={'label': 'Average Total Charges ($)'})
plt.title('Average Total Charges by Ethnicity and Race')
plt.xlabel('Race')
plt.ylabel('Ethnicity')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()

# Step 3: Create a Heatmap for Average Length of Stay by Ethnicity and Race
plt.figure(figsize=(12, 8))
length_of_stay_pivot = ethnicity_race_analysis.pivot("Ethnicity", "Race", "Average_Length_of_Stay")
sns.heatmap(length_of_stay_pivot, annot=True, fmt=".2f", cmap="YlGnBu", cbar_kws={'label': 'Average Length of Stay (days)'})
plt.title('Average Length of Stay by Ethnicity and Race')
plt.xlabel('Race')
plt.ylabel('Ethnicity')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()

# Step 4: Pairplot to explore relationships between Total Charges and Length of Stay
sns.pairplot(filtered_df_cleaned, vars=['Total_Charges', 'Length_of_Stay'], hue='Ethnicity', palette='husl')
plt.suptitle('Pairplot of Total Charges and Length of Stay by Ethnicity', y=1.02)
plt.show()
```
